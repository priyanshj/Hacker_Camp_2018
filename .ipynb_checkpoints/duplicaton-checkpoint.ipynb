```python


def preProcess(column):
    try : # python 2/3 string differences
       column = column.decode('utf8')
    except AttributeError:
       pass
    column = unidecode(column)
    column = re.sub('  +', ' ', column)
    column = re.sub('\n', ' ', column)
    column = column.strip().strip('"').strip("'").lower().strip()



#If data is missing, indicate that by setting the value to None

    if not column:
        column = None
    return column



df2 = pd.read_csv("real.csv")
df2.drop(["Unnamed: 0"],axis=1,inplace=True)
df2.head()
df2.to_csv("max.csv")
df2.head()
```

```python
#Function to read file

def readData(filename):
    data_d = {}
    with open(filename) as f:
        reader = csv.DictReader(f)
        for row in reader:
            clean_row = [(k,preProcess(v))for (k, v) in row.items()]
            row_id = (row['Id'])
            data_d[row_id] = dict(clean_row)

    return data_d
```

```python
#Defining all files
input_file = 'max.csv'
output_file = 'csv_example_output.csv'
settings_file = 'csv_example_learned_settings'
training_file = 'csv_example_training.json'

```

```python
print('importing data ...')
data_d = readData(input_file)

#If a settings file already exists, we'll just load that and skip training

if os.path.exists(settings_file):
    print('reading from', settings_file)
    with open(settings_file, 'rb') as f:
        deduper = dedupe.StaticDedupe(f)
else:
    fields = [
        {'field' : 'dob', 'type': 'String'},
        {'field' : 'gn', 'type': 'String'},
        {'field' : 'fn', 'type': 'String'},
        {'field' : 'ln', 'type': 'String'}
        ]

#Training
```

```python
#Defining Fields

fields = [
        {'field' : 'dob', 'type': 'String'},
        {'field' : 'gn', 'type': 'String'},
        {'field' : 'fn', 'type': 'String'},
        {'field' : 'ln', 'type': 'String'}
        ]

```

```python
deduper = dedupe.Dedupe(fields)
deduper.sample(data_d, 200)
              


if os.path.exists(training_file):
    print('reading labeled examples from ', training_file)
    with open(training_file, 'rb') as f:
        deduper.readTraining(f)


#Active learning

#Dedupe will find the next pair of records it is least certain about and ask to label them as duplicates or not. 

print('starting active labeling...')

dedupe.consoleLabel(deduper)
```

```python
deduper.train()

with open(training_file, 'w') as tf:
        deduper.writeTraining(tf)

with open(settings_file, 'wb') as sf:
        deduper.writeSettings(sf)
        



threshold = deduper.threshold(data_d, recall_weight=1)


        
```

```python
#clustering the data
print('clustering...')
clustered_dupes = deduper.match(data_d, threshold)
clustered_dupes_2 = deduper.match(data_d)
print('# duplicate sets', len(clustered_dupes))

```
